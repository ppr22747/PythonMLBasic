#### 1. Simple Linear Regression (y = ax +b)

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/Advertising1.csv')
df
```

```python
df.corr()
```

```python
df.plot(x='TV',y='Sales',style='o')
plt.ylabel('Sales')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = df[['TV']]
y = df['Sales']
```

```python
lrm = LinearRegression()
lrm.fit(X, y)
```

```python
lrm.intercept_ # b from y = ax+b
```

```python
lrm.coef_ # a from y = ax+b
```

```python
predictions = lrm.predict(X)
predictions
```

```python
plt.scatter(X,y,color='black') # scatter
plt.plot(X,predictions,color='blue',linewidth=3) # line
```

```python
lrm.predict([[130]])
```

```python
lrm.predict([[500]])
```

```python
from sklearn import metrics
metrics.mean_absolute_error(y, predictions) # find MAE
```

#### 2. Multiple Linear Regression y = b0 + b1x1 + b2x2 + ... + bixi + E

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/Advertising2.csv')
df
```

```python
df.corr()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = df[['TV', 'Radio', 'Newspaper']]
y = df['Sales']
```

```python
lrm = LinearRegression()
lrm.fit(X, y)
```

```python
predictions = lrm.predict(X)
predictions
```

```python
from sklearn import metrics
metrics.mean_absolute_error(y, predictions)
```

```python
lrm.predict([[10,300,40]])
```

```python
lrm.predict([[300,20,10]])
```
