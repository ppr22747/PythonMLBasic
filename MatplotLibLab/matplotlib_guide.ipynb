#### 2. create Line plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# define x and y axis with numpy array

x = np.array([2018,2019,2020,2021,2022])
y = np.array([1050,2454,3512,4652,4052])

plt.plot(x,y)
plt.xticks(x) # turn x to text
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

#### 3. create bar plot

```python
plt.bar(x,y)
plt.xticks(x) # turn x to text
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

#### 4. create plot from csv

```python
import matplotlib.pyplot as plt
import pandas as pd

# create dataframe from csv

df = pd.read_csv('datasets/gdp.csv')
# create line plot
plt.plot(df['year'],df['gdp'])
plt.xticks(df['year'])
plt.title('Thailand GDP')
plt.xlabel('Year')
plt.ylabel('GDP (Billion THB)')
plt.show
```

```python
# create bar chart
plt.bar(
    df['year'], df['gdp'],
    color=['green','red','blue','yellow','pink'],
    edgecolor='black'
)

plt.xticks(df['year'])
plt.title('Thailand GDP')
plt.xlabel('Year')
plt.ylabel('GDP (Billion THB)')
plt.show
```

```python
# create pie chart

# create bar chart
plt.pie(
    df['gdp'],
    labels=df['year'],
    autopct='%.2f%%',
    colors=['green','red','blue','yellow','pink']
)

plt.title('Thailand GDP')
plt.show
```

```python
# scatter plot

import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('datasets/gdp.csv')

plt.scatter(df['year'],df['gdp'])
plt.xticks(df['year'])
plt.title('Thailand GDP')
plt.xlabel('Year')
plt.ylabel('GDP (Billion THB)')
plt.show()
```
